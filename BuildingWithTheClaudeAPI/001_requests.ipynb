```python
# Careate an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
def add_user_message(messages, text):  
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


def add_assistant_message(messages, text):  
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)




def chat(messages):
  message = client.messages.create( 
    model=model,
    max_tokens=1000,
    messages=messages,
  )
  return message.content[0].text
```

```python
# Make a starting list of messages
messages = []
# Add in the initial user question of "Define quantum computing in one sentence"
add_user_message(messages, "Define quantum computing in one sentence")

# Pass the list of messages into 'chat' to get the answer
answer=chat(messages)

# Take the answer and add it to the list of messages
add_assistant_message(messages, answer)

# Add in the user's follow-up question
add_user_message(messages, "write another sentence")
# Call chat again with the list of messages to get a final answer
answer=chat(messages)

print("Claude:----------")
print(answer)
print("----------")
# add_assistant_message(messages, chat(messages))



```

```python
# prompt the user to enter some input using the built-in input function

messages1 = []

while True:
    user_input = input("Enter some input: ")
    if user_input.strip().lower() in {"exit", "quit", "q"}:
        break
    # add the user's input to the list of messages
    add_user_message(messages1, user_input)

    # call the chat function with the list of messages
    answer=chat(messages1)  

    # add the assistant's response to the list of messages
    add_assistant_message(messages1, answer)

    # print the final list of mess1ages
    print("Claude:----------")
    print(f"Claude: {answer}\n")
    print("----------")
print("done")

# repeat from #1
```
